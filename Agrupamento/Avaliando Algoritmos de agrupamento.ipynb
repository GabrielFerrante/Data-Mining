```python
!pip install plotly --upgrade
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
x_random, y_random = datasets.make_moons(n_samples = 1500, noise = 0.09)
```

```python
x_random
```

```python
grafico = px.scatter(x = x_random[:,0], y = x_random[:,1])
grafico
```

```python
#TESTANDO O KMEANS

agrupador = KMeans(n_clusters = 2)
rotulos = agrupador.fit_predict(x_random)
grafico = px.scatter(x = x_random[:,0], y = x_random[:,1], color = rotulos)
grafico.show()
```

```python
#TESTANDO HIERARQUICO
agrupador = AgglomerativeClustering(n_clusters = 2, metric= 'euclidean', linkage = 'ward')
rotulos = agrupador.fit_predict(x_random)
grafico = px.scatter(x = x_random[:,0], y = x_random[:,1], color = rotulos)
grafico.show()
```

```python
#TESTANDO O DBSCAN

agrupador = DBSCAN(eps = 0.1)
rotulos = agrupador.fit_predict(x_random)
grafico = px.scatter(x = x_random[:,0], y = x_random[:,1], color = rotulos)
grafico.show()
```
