```python
!pip install plotly --upgrade
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
```

```python
base1 = pd.read_csv('plano_saude2.csv')
X = base1.iloc[:, 0:1].values
Y = base1.iloc[:, 1].values
```

```python
X
```

```python
#KERNEL LINEAR

regressor = SVR(kernel = 'linear')
regressor.fit(X,Y)
regressor.score(X,Y) #Coeficiente de determinação
```

```python
grafico = px.scatter(x = X.ravel(), y = Y) #GRAFICO DE DISPERSÃO
grafico.add_scatter(x = X.ravel(), y = regressor.predict(X), name = 'Regressão Arvore')
grafico.show()
```

```python
#KERNEL POLINOMIAL

regressor = SVR(kernel = 'poly', degree=3) #degree - Versao original, ao quadrado e ao  cubo
regressor.fit(X,Y)
regressor.score(X,Y)
```

```python
grafico = px.scatter(x = X.ravel(), y = Y) #GRAFICO DE DISPERSÃO
grafico.add_scatter(x = X.ravel(), y = regressor.predict(X), name = 'Regressão Arvore')
grafico.show()
```

```python
#KERNEL RBF
novoX = StandardScaler() #COLOCANDO TUDO NA MESMA ESCALA
X = novoX.fit_transform(X)
novoY = StandardScaler()
Y = novoY.fit_transform(Y.reshape(-1,1))


regressor = SVR(kernel = 'rbf')
regressor.fit(X,Y.ravel())
regressor.score(X,Y) #Coeficiente de determinação
```

```python
grafico = px.scatter(x = X.ravel(), y = Y.ravel()) #GRAFICO DE DISPERSÃO
grafico.add_scatter(x = X.ravel(), y = regressor.predict(X), name = 'Regressão Arvore')
grafico.show()
```

```python
base = pd.read_csv('house_prices.csv')
base.drop('date', axis='columns', inplace=True) #REMOVENDO A COLUNA EM FORMATO DE DATA
X_casas = base.iloc[:,2:18].values
y_casas = base.iloc[:,1].values
XcasasTreino, XcasasTeste, YcasasTreino, YcasasTeste = train_test_split(X_casas,y_casas, test_size = 0.3, random_state=0)
XcasasTreino,YcasasTreino
```

```python
scalerX = StandardScaler()
XcasasTreino = scalerX.fit_transform(XcasasTreino)
scalerY = StandardScaler()
YcasasTreino = scalerY.fit_transform(YcasasTreino.reshape(-1,1))
XcasasTreino,YcasasTreino
```

```python
XcasasTeste = scalerX.transform(XcasasTeste)
YcasasTeste = scalerY.transform(YcasasTeste.reshape(-1,1))
XcasasTeste,YcasasTeste
```

```python
regressor = SVR(kernel='rbf')

regressor.fit(XcasasTreino,YcasasTreino)
```

```python
regressor.score(XcasasTreino,YcasasTreino)
```

```python
regressor.score(XcasasTeste,YcasasTeste)
```

```python
previsoes = regressor.predict(XcasasTeste)
previsoes
```

```python
YcasasTeste
```

```python
inversoY = scalerY.inverse_transform(YcasasTeste)
inversoY
```

```python
invernoPrevisoes = scalerY.inverse_transform(previsoes.reshape(-1,1))
invernoPrevisoes
```

```python
mean_absolute_error(inversoY,invernoPrevisoes) #pode errar 82 mil dolares para cima ou para baixo
```
