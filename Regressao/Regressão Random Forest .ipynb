```python
!pip install plotly --upgrade
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
base1 = pd.read_csv('plano_saude2.csv')
X = base1.iloc[:, 0:1].values
Y = base1.iloc[:, 1].values

regressor = RandomForestRegressor(n_estimators=10)
regressor.fit(X,Y)
```

```python
regressor.score(X,Y)
```

```python
X_teste_arvore = np.arange(min(X), max(X), 0.1) #GERANDO UM DATASET DE TESTE FALSA
X_teste_arvore = X_teste_arvore.reshape(-1,1)
X_teste_arvore.shape
```

```python
grafico = px.scatter(x = X.ravel(), y = Y) #GRAFICO DE DISPERSÃO
grafico.add_scatter(x = X_teste_arvore.ravel(), y = regressor.predict(X_teste_arvore), name = 'Regressão Arvore')
grafico.show()
```

```python
base = pd.read_csv('house_prices.csv')
base.drop('date', axis='columns', inplace=True) #REMOVENDO A COLUNA EM FORMATO DE DATA
X_casas = base.iloc[:,2:18].values
y_casas = base.iloc[:,1].values
XcasasTreino, XcasasTeste, YcasasTreino, YcasasTeste = train_test_split(X_casas,y_casas, test_size = 0.3, random_state=0)
```

```python
XcasasTreino.shape, YcasasTreino.shape
```

```python
regressor = RandomForestRegressor(n_estimators=100)
regressor.fit(XcasasTreino,YcasasTreino)
```

```python
regressor.score(XcasasTreino,YcasasTreino)
```

```python
regressor.score(XcasasTeste,YcasasTeste)
```

```python
previsoes = regressor.predict(XcasasTeste)
```

```python
mean_absolute_error(YcasasTeste, previsoes) #MAE #Diferenças reais
```
