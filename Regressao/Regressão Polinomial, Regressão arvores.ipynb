```python
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
base = pd.read_csv('plano_saude2.csv')
```

```python
base
```

```python
X = base.iloc[:, 0:1].values
X
```

```python
Y = base.iloc[:, 1].values
Y
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree = 4) #numero de vezes que os atributos serão replicados e elevados TESTE COM OUTROS DEGRAIS
X_poly = poly.fit_transform(X)
```

```python
X.shape
```

```python
18 * 18
```

```python
X_poly #segundo valor é o valor original, e o terceiro é o quadrado dele.
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
```

```python
regressor.fit(X_poly, Y)
```

```python
regressor.intercept_  #B0
```

```python
regressor.coef_ #Bn's
```

```python
novo = poly.transform([[40]])
novo
```

```python
regressor.predict(novo)
```

```python
previsoes = regressor.predict(X_poly)
previsoes
```

```python
grafico = px.scatter(x = X.ravel(), y = Y) #GRAFICO DE DISPERSÃO
grafico.add_scatter(x = X.ravel(), y = previsoes, name = 'Regressão Poly')
grafico.show()
```

BASE DAS CASAS

```python
base = pd.read_csv('house_prices.csv')
```

```python
base.drop('date', axis='columns', inplace=True) #REMOVENDO A COLUNA EM FORMATO DE DATA
```

```python
Xcasas = base.iloc[:,2:18].values
Ycasas = base.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split

XcasasTreino, XcasasTeste, YcasasTreino, YcasasTeste = train_test_split(Xcasas,Ycasas, test_size = 0.3, random_state=0)
```

```python
XcasasTreino.shape, YcasasTreino.shape
```

```python
XcasasTeste.shape, YcasasTeste.shape
```

```python
poly = PolynomialFeatures(degree = 2)
XpolyTreino = poly.fit_transform(XcasasTreino)
XpolyTeste = poly.fit_transform(XcasasTeste)
```

```python
XpolyTreino.shape, XpolyTeste.shape
```

```python
regressor = LinearRegression()

regressor.fit(XpolyTreino,YcasasTreino)
```

```python
regressor.score(XpolyTreino,YcasasTreino)
```

```python
regressor.score(XpolyTeste,YcasasTeste)
```

```python
previsoes = regressor.predict(XpolyTeste)
previsoes
```

```python
YcasasTeste
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(YcasasTeste, previsoes) #MAE #Diferenças reais
```

ARVORES DE REGRESSÃO

BASE DE PLANO DE SAUDE

```python
from sklearn.tree import DecisionTreeRegressor

base = pd.read_csv('plano_saude2.csv')
X = base.iloc[:, 0:1].values
Y = base.iloc[:, 1].values
regressorArvore = DecisionTreeRegressor()

regressorArvore.fit(X,Y)
regressorArvore.score(X,Y)
```

```python
previsoes = regressorArvore.predict(X)
previsoes
```

```python
grafico = px.scatter(x = X.ravel(), y = Y) #GRAFICO DE DISPERSÃO
grafico.add_scatter(x = X.ravel(), y = previsoes, name = 'Regressão Arvore')
grafico.show()
```

```python
X_teste_arvore = np.arange(min(X), max(X), 0.1) #GERANDO UM DATASET DE TESTE FALSO
X_teste_arvore
```

```python
X_teste_arvore = X_teste_arvore.reshape(-1,1)
X_teste_arvore.shape
```

```python
grafico = px.scatter(x = X.ravel(), y = Y) #GRAFICO DE DISPERSÃO
grafico.add_scatter(x = X_teste_arvore.ravel(), y = regressorArvore.predict(X_teste_arvore), name = 'Regressão Arvore')
grafico.show()
```
