```python
import pandas as pd

base = pd.read_csv('census.csv')

base
```

```python
colunas = base.columns[:-1]
colunas
```

```python
X_census = base.iloc[:, 0:14].values
Y_census = base.iloc[:, 14].values
```

```python
X_census
```

```python
Y_census
```

```python
#LABEL ENCODER

from sklearn.preprocessing import LabelEncoder
label_encoder_workclass = LabelEncoder()
label_encoder_education = LabelEncoder()
label_encoder_marital = LabelEncoder()
label_encoder_occupation = LabelEncoder()
label_encoder_relationship = LabelEncoder()
label_encoder_race = LabelEncoder()
label_encoder_sex = LabelEncoder()
label_encoder_country = LabelEncoder()

X_census[:,1] = label_encoder_workclass.fit_transform(X_census[:,1])
X_census[:,3] = label_encoder_education.fit_transform(X_census[:,3])
X_census[:,5] = label_encoder_marital.fit_transform(X_census[:,5])
X_census[:,6] = label_encoder_occupation.fit_transform(X_census[:,6])
X_census[:,7] = label_encoder_relationship.fit_transform(X_census[:,7])
X_census[:,8] = label_encoder_race.fit_transform(X_census[:,8])
X_census[:,9] = label_encoder_sex.fit_transform(X_census[:,9])
X_census[:,13] = label_encoder_country.fit_transform(X_census[:,13])
```

```python
X_census
```

```python
#Normalização

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler() #Pode ser também a padronização StandardScaler() com DesvioPadrao se tivesse Outliers na base
Xescalado = scaler.fit_transform(X_census)
Xescalado
```

LOW VARIANCE - BASEADO EM MEDIDA DE DISPERSÃO ESTATÍSTICA, COM BASE EM VARIANCIA. LEMBRANDO, É A DISTANCIA DA MÉDIA

```python
for item in range(X_census.shape[1]): #SHAPE [1] colunas
    print(Xescalado[:, item].var()) #extração da variancia para cada atributo
```

```python
from re import X
from sklearn.feature_selection import VarianceThreshold
#LIMIAR DE VARIANCIA
#NOTE QUE VALORES BAIXOS, SÃO VALORES QUE SE APROXIMAM DA MÉDIA
#BUSCAMOS VALORES ALTOS NESTE CASO
selecao = VarianceThreshold(threshold=0.05)
X_census_variancia = selecao.fit_transform(Xescalado)
X_census_variancia.shape
```

```python
X_census_variancia #Colunas reduzidas, devido a escolha baseada em variancia
```

```python
selecao.variances_ #SAO AS MESMAS VARIANCIAS
```

```python
import numpy as np
indices = np.where(selecao.variances_ > 0.05)
indices #MOSTRANDO QUAIS ATRIBUTOS FORAM SELECIONADOS
```

```python
colunas[indices]
```

```python
baseNova = base.drop(columns = ['age', 'workclass', 'final-weight', 'education-num',
                                'race', 'capital-gain', 'capital-loos', 'hour-per-week', 'native-country'], axis = 1)
```

```python
baseNova
```

```python
treinoAtributos = baseNova.iloc[:, 0:5].values
testeAtributos = baseNova.iloc[:, 5].values
```

```python
treinoAtributos
```

```python
testeAtributos
```

```python
treinoAtributos[:, 0] = label_encoder_education.fit_transform(treinoAtributos[:,0])
treinoAtributos[:, 1] = label_encoder_marital.fit_transform(treinoAtributos[:,1])
treinoAtributos[:, 2] = label_encoder_occupation.fit_transform(treinoAtributos[:,2])
treinoAtributos[:, 3] = label_encoder_relationship.fit_transform(treinoAtributos[:,3])
treinoAtributos[:, 4] = label_encoder_sex.fit_transform(treinoAtributos[:,4])
```

```python
treinoAtributos
```

```python
#ONE HOT ENCODER
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

onehot = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [0,1,2,3,4])], remainder='passthrough')

treinoAtributos = onehot.fit_transform(treinoAtributos).toarray()

treinoAtributos
```

```python

scaler = MinMaxScaler() #Pode ser também a padronização StandardScaler() com DesvioPadrao se tivesse Outliers na base
treinoAtributosEscalado = scaler.fit_transform(treinoAtributos)
treinoAtributosEscalado
```

```python
#APLICANDO O SPLIT

from sklearn.model_selection import train_test_split

#X se refere ao atributos previsores
#Y se refere ao atributo alvo da classificação
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(treinoAtributosEscalado, testeAtributos, test_size= 0.25, random_state=0)
```

```python
X_treinamento.shape, X_teste.shape #SEM ESSA TECNICA, TINHAMOS ANTERIORMENTE 108
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_credit = RandomForestClassifier(n_estimators = 40, criterion = 'entropy', random_state = 0)

random_forest_credit.fit(X_treinamento,y_treinamento)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

previsoes = random_forest_credit.predict(X_teste)
accuracy_score(y_teste, previsoes)
```

```python
importancias.sum()
```

```python
indices = []

for i in range(len(importancias)):
  if importancias[i] >= 0.029:
    indices.append(i)

indices
```

```python
colunas[indices]
```

```python
XExtra = X_census[:, indices]

XExtra
```

```python
#ONE HOT ENCODER
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

onehot = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [1,3,5,6,7])], remainder='passthrough')

XExtra = onehot.fit_transform(XExtra).toarray()

XExtra
```

```python
XExtra.shape
```

```python
#APLICANDO O SPLIT

from sklearn.model_selection import train_test_split

#X se refere ao atributos previsores
#Y se refere ao atributo alvo da classificação
X_treinamentoExtra, X_testeExtra, y_treinamentoExtra, y_testeExtra = train_test_split(XExtra, testeAtributos, test_size= 0.25, random_state=0)
X_treinamentoExtra.shape, X_testeExtra.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_credit = RandomForestClassifier(n_estimators = 40, min_samples_leaf=1, min_samples_split=5, criterion = 'entropy', random_state = 0)

random_forest_credit.fit(X_treinamentoExtra,y_treinamentoExtra)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

previsoes = random_forest_credit.predict(X_testeExtra)
accuracy_score(y_testeExtra, previsoes)
```
