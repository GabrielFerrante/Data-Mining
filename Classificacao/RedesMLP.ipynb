```python
from sklearn.neural_network import MLPClassifier
```

```python
import pickle
with open('credit.pkl', 'rb') as f:
  X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste = pickle.load(f)
```

```python
X_credit_treinamento.shape, y_credit_treinamento.shape
```

```python
X_credit_teste.shape, y_credit_teste.shape
```

```python
(3 + 1) / 2
```

```python
# REDE 3 -> 20 -> 20 -> 1

#max_iter = épocas
#verbose = mostrar o valor do erro pelas epocas
#tol = tolencia, ou seja, o valor de redução do erro por epoca
#sover = otimizador da atualização dos pesos, pode ser SGD, ADAM
#activation = função de ativação
#hidden layer sizes = numero de neuronios por camada. Ex: 100, 100 - indica duas camadas com 100 neuronios cada
rede_neural_credit = MLPClassifier(max_iter=1500, verbose=True, tol=0.0000100,
                                   solver = 'adam', activation = 'relu',
                                   hidden_layer_sizes = (20,20))
rede_neural_credit.fit(X_credit_treinamento, y_credit_treinamento)
```

```python
previsoes = rede_neural_credit.predict(X_credit_teste)
previsoes
```

```python
y_credit_teste
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_credit_teste, previsoes)
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(rede_neural_credit)
cm.fit(X_credit_treinamento, y_credit_treinamento)
cm.score(X_credit_teste, y_credit_teste)
```

```python
print(classification_report(y_credit_teste, previsoes))
```
